```
from pyspark.ml.feature import VectorAssembler
```

```
dataset = spark.table('kaggle_titanic.train_selected')
```

```
features = ['Pclass', 'Sex_index']
assembler = VectorAssembler(inputCols=features, outputCol='features')
dataset_assembled = assembler.transform(dataset)
```

```
display(dataset_assembled)
```

```
### Randomly split data into training and test sets. set seed for reproducibility
(train, test) = dataset_assembled.randomSplit([0.7, 0.3], seed=100)
print("Training data size: {0}".format(train.count()))
print("Testing data size: {0}".format(test.count()))
```

```
from pyspark.ml.classification import LogisticRegression
# Create initial LogisticRegression model
logistic_regression = LogisticRegression(labelCol="Survived", featuresCol="features", maxIter=10)

# Train model with Training Data
model = logistic_regression.fit(train)
```

```
predictions_train = model.transform(train)
predictions_test = model.transform(test)
```

```
# calculate train accuracy
train_results = predictions_train.groupBy('Survived').pivot('prediction').count()
train_results.show()
train_results = train_results.collect()
print("Train Accuracy: {0}".format((train_results[0][2]+train_results[1][1])/train.count()))
```

```
# calculate test accuracy
test_results = predictions_test.groupBy('Survived').pivot('prediction').count()
test_results.show()
test_results = test_results.collect()
print("Testing Accuracy: {0}".format((test_results[0][2]+test_results[1][1])/test.count()))
```
